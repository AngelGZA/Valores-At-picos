```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
data
```

```python
#Realizamos una copia del dataframe
data2=data.copy()
```

```python
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos para dataframe
valores_nulos=data2.isnull().sum().sum()
valores_nulos
```

```python
#Cuarto metodo de sustitucion de valores nulos
#Sustituir valores nulos por un string en concreto
data2["NOMBRE_VENDEDOR"]=data2["NOMBRE_VENDEDOR"].fillna("Unknown")
```

```python
#Identificar valores nulos para dataframe
valores_nulos=data2.isnull().sum().sum()
valores_nulos
```

```python
#Verificamos informacion del DataFrame
data2.info()
```

```python
#Creo 2 dataframe para poder procesar los outliers
cuantitativas= data2.iloc[ : , 7:13]
cualitativas= data2.iloc[ : , 0: 6 ]
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize=(15,8))
cuantitativas.plot(kind='box', vert=False)
plt.title("Valores Atipicos del DataFrame")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO "DESVIACION ESTANDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Metodo aplicando desviacion estandar. Encuentro los valores extremos
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Reemplazamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Metodo aplicando Cuartiles. Encuentro cuartiles 0:25 y 0:75
y=cuantitativas

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr = percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con "mean"
#Reemplazamos una copia del dataframe
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
#Unimos el dataframe cuantitativos limpios con el dataframe cualitativo
Datos_limpios = pd.concat([cualitativas, data4_iqr], axis=1)
Datos_limpios
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame
Datos_limpios.to_csv("Datos_limpios_Ejercicio_4.1.csv")
```
